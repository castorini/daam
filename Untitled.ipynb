```python
from daam import trace, set_seed, plot_overlay_heat_map, expand_image
from diffusers import StableDiffusionPipeline
from matplotlib import pyplot as plt
import torch
```

```python
model_id = 'CompVis/stable-diffusion-v1-4'
device = 'cuda'

pipe = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True)
pipe = pipe.to(device)
```

```python
prompt = 'Two dogs are running in sunlight'
for layer_i in [1]:
    for head_i in range(2):
        print('[head index]: ', head_i)
        gen = set_seed(811)
        with torch.cuda.amp.autocast(dtype=torch.float16), torch.no_grad():
                with trace(pipe, layer_idx=layer_i, head_idx=head_i) as tc:
                    out = pipe(prompt, num_inference_steps=30, generator=gen)
                    heat_map = tc.compute_global_heat_map(prompt)
                    heat_map = expand_image(heat_map.compute_word_heat_map('dogs'))
                    plot_overlay_heat_map(out.images[0], heat_map)
                    plt.show()
```
